```python
def solution(park, routes):
    max_bottom = len(park) - 1
    max_left = len(park[0]) - 1    
    cur_loc = [0, 0]
    
    for height, row in enumerate(park):
        for width, ch in enumerate(row):
            if ch == 'S':
                cur_loc[0] = height
                cur_loc[1] = width    
    
    for route in routes:
        direction = route[0]        
        move = int(route[-1])
        
        safe_flag = True
        if direction == 'N':
            new_y = cur_loc[0] - move
            if (new_y < 0) or (max_bottom < new_y):
                continue
                
            for row in park[new_y:cur_loc[0]+1]:
                if row[cur_loc[1]] == 'X':
                    safe_flag = False
                    break
            if not safe_flag:
                continue
                
            target_loc = (new_y, cur_loc[1])
            
        elif direction == 'S':
            new_y = cur_loc[0] + move
            if (new_y < 0) or (max_bottom < new_y):
                continue
            
            for row in park[cur_loc[0]: new_y+1]:
                if row[cur_loc[1]] == 'X':
                    safe_flag = False
                    break
            if not safe_flag:
                continue                      
            
            target_loc = (new_y, cur_loc[1])
            
        elif direction == 'W':
            new_x = cur_loc[1] - move
            if (new_x < 0) or (max_left < new_x):
                continue
                
            for ch in park[cur_loc[0]][new_x:cur_loc[1]+1]:
                if ch == 'X':
                    safe_flag = False
                    break
            
            if not safe_flag:
                continue
            
            target_loc = (cur_loc[0], new_x)
            
        elif direction == 'E':
            new_x = cur_loc[1] + move
            if (new_x < 0) or (max_left < new_x):
                continue
                
            for ch in park[cur_loc[0]][cur_loc[1]: new_x+1]:
                if ch == 'X':
                    safe_flag = False
                    break
                    
            if not safe_flag:
                continue
            target_loc = (cur_loc[0], new_x)            
        
        cur_loc = target_loc       
        
    return cur_loc
```
