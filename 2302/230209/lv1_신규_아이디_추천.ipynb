```python
def process_stage_1(new_id):
    return ''.join([ch.lower() for ch in new_id])

def process_stage_2(stage_1):
    stage_2 = ''
    for ch in stage_1:
        if ord('a') <= ord(ch) and ord(ch) <= ord('z'):
            stage_2 += ch
        elif ord('0') <= ord(ch) and ord(ch) <= ord('9'):
            stage_2 += ch
        elif ord(ch) in [ord('.'), ord('-'), ord('_')]:
            stage_2 += ch
    return stage_2

def process_stage_3(stage_2):
    stage_3 = ''
    for ch in stage_2:
        if ch != '.':
            stage_3 += ch
        elif ch == '.':
            if stage_3 == '':
                stage_3 += ch
            else:
                if stage_3[-1] == '.':
                    continue
                stage_3 += ch
    return stage_3        

def process_stage_4(stage_3):
    stage_4 = stage_3
    if stage_3[0] == '.':
        stage_4 = stage_4[1:]
    if stage_3[-1] == '.':
        stage_4 = stage_4[:-1]
    return stage_4

def process_stage_567(stage_4):
    answer = stage_4
    if answer == '':
        answer = 'a'
    if len(answer) >= 16:
        answer = answer[:15]
    if answer[-1] == '.':
        answer = answer[:-1]
    if len(answer) <= 2:
        while True:
            if len(answer) >= 3:
                break            
            answer += answer[-1]
    return answer

def solution(new_id):
    stage_1 = process_stage_1(new_id)
    stage_2 = process_stage_2(stage_1)
    stage_3 = process_stage_3(stage_2)
    stage_4 = process_stage_4(stage_3)
    answer = process_stage_567(stage_4)
    return answer
```
